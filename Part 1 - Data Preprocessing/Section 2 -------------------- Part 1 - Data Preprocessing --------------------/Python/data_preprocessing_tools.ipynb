# Data Preprocessing Tools

## Importing the libraries

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```
dt = pd.read_csv ('Data.csv')
x = dt.iloc[:,: -1].values
y = dt.iloc[:,-1].values
```

## Taking care of missing data

```
from sklearn.impute import SimpleImputer
#Menghitung jumlah data yang kosong dan menampilkannya
  # nun = pd.isnull().sum()
# Kita membuat objek SimpleImputer yang akan mencari nilai yang hilang (np.nan) dan menggantinya dengan nilai rata-rata dari kolom yang bersangkutan.
imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
# Kita melatih imputer dengan data dari kolom ke-2 dan ke-3 (x[:, 1:3]). Ini berarti imputer menghitung rata-rata untuk setiap kolom tersebut.
imputer.fit(x[:,1:3])
# Kita menggantikan nilai yang hilang pada kolom ke-2 dan ke-3 dengan rata-rata yang sudah dihitung.# Data dari kolom tersebut diimputasi (diisi) oleh imputer dan hasilnya ditugaskan kembali ke kolom yang sama di x.
x[:, 1:3] = imputer.transform(x[:,1:3])
```

```
print(x)
```

## Encoding categorical data

### Encoding the Independent Variable

Merubah isi **kolom negara dengan 0 dan 1**. **Dari pada harus jadi urutan numerik 0 higga 2**. Itu akan **membuat model ML mengira ada urutan numerik antara negara.**

```
# Melakukan perubahan pada data kategorikal, menjadi numerik.

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# 1. encoder = nama bebas yang akan merujuk pada transformasi nanti,
# 2. OneHotEncoder = jenis transformasi nya yaitu merubah data katagorikal menjadi numerik
# 3. remainder 'passthrough' = Mengingatkan agar kolom yang tidak di pilih (Dibiarkan saja dan dilewatkan )
# 4. [0] = Memilih kolom pertama yg akan di lakukan

ct = ColumnTransformer(transformers = [('encoder', OneHotEncoder(),[0])],remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```
print (x)
```

### Encoding the Dependent Variable

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```
print (y)
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split (x , y, test_size = 0.2, random_state = 1)
```
