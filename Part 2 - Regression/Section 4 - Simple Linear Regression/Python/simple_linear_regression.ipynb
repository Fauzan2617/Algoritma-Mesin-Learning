# Simple Linear Regression

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dt = pd.read_csv ('Salary_Data.csv')
X = dt.iloc[:,:-1].values
y = dt.iloc [:,-1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,text_size = 1/3, random_state = 1)
```

## Training the Simple Linear Regression model on the Training set

```python
# REGRESSION KLASIFIKASI (Model untuk menetunkan data yg tidak berkelanjutan)
# REGRESSION LINEAR (Model menentukan data yang berkelanjutan atau kesinambungan)
from sklearn.linear_model import LinearRegression
regressi = LinearRegression ()
regressi.fit(X_train,y_train)
```

* Model Model ML bukan hanya memiliki metode TETAPI ALAT (Alat untuk hasil pengujian, melatih model dan prediksi)

## Predicting the Test set results

```python
y_predict = regressi.predict(X_test)
```

## Visualising the Training set results

```python
plt.scatter(X_train,y_train, color = 'red')
plt.plot(X_train, regressi.predict(X_train),color = 'blue')
plt.title ('Pengalaman vs Gajih')
plt.xlabel('pengalaman')
plt.ylabel('gajih')
plt.show()
```

## Visualising the Test set results

```python
plt.scatter(X_test,y_test, color = 'red')
plt.plot(X_train, regressi.predict(X_train),color = 'blue')
plt.title ('Pengalaman vs Gajih')
plt.xlabel('pengalaman')
plt.ylabel('gajih')
plt.show()
```

* Prediksi dalam 12 tahun pengalaman

```python
print (regressi.predict([[12]]))
```

Dalam regressi memiliki persamaan:
  y = mX + b
-m = memprediksi output y dengan input x jika bertambah 1. Atau ini menunjukkan kemiringan garis.
-b = koefisien b dalam memprediksi output y ketika x input 0

```python
print (regressi.coef_)
print (regressi.intercept_)
```
